```python
import random
```

```python
def get_n_random_numbers(n=10, min_=-5, max_=5):
    numbers = []
    for i in range(n):
        numbers.append(random.randint(min_,max_))
    return numbers
get_n_random_numbers()
```

```python
def my_linear_search(list, item_search):
    found = (-1,-1) #item_search listede yoksa. 
    n = len(list)
    for indis in range(n):
        if list[indis] == item_search: #Listede item_search var mı?
            found = (list[indis], indis)
            break #Listede bulduğun anda çık.
    return found
```

```python
def my_experimental_study_linear(iterNum = 50):
    cost = []
    x_low = -100
    x_high = 100
    array_size = 40
    print("Array size : ",array_size)
    for iter in range(iterNum):
        my_list = get_n_random_numbers(array_size, x_low, x_high)
        my_search_item = get_n_random_numbers(1, x_low, x_high)
        my_search_item = my_search_item[0]
        result = my_linear_search(my_list, my_search_item)
        print(my_list)
        print(my_search_item)
        if(result[1] == -1):
            cost.append(array_size)
        else:
            cost.append(result[1])
        print(result)
    #my_list, my_search_item
    return cost
```

```python
def binary_search(list, item_search):
    found = (-1,-1,0) #Aranan değer, indis, kaç defa search yaptım. 
    low = 0
    high = len(list) - 1
    s = 0
    while low<=high:
        mid = (low + high) // 2
        print(low, high, mid)
        s = s + 1
        if list[mid] == item_search:
            return list[mid], mid, s
        elif list[mid] > item_search: 
            high = mid - 1
        else:
            low = mid + 1
    print(s)
    return found[0], found[1], s #None
```

```python
def bubble_sort(list):
    n = len(list)
    for i in range(n-1, -1, -1):
        for j in range(0, i):
            if not(list[j] < list[j+1]):
                t = list[j]
                list[j] = list[j + 1]
                list[j + 1] = t
    return list
```

```python
def my_experimental_study_binary(iterNum = 50):
    cost = []
    x_low = -100
    x_high = 100
    array_size = 40
    print("Array size : ",array_size)
    for iter in range(iterNum):
        my_list = get_n_random_numbers(array_size, x_low, x_high)
        my_list = bubble_sort(my_list)
        my_search_item = get_n_random_numbers(1, x_low, x_high)
        my_search_item = my_search_item[0]
        result = binary_search(my_list, my_search_item)
        cost.append(result[2])
    return cost
```
